## Reading Data

### Note :
#### We can pass python data structures to create dataframe
#### Using the columns parameter allows us to tell the constructor , how we will like our column ordered. By default , the data frame constructor orders column alphabetically ( though this isn't the case when reading from a file )

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option('max_columns', 50)
%matplotlib inline
```

```python
data={'year':[2010,2011,2012,2011,2012,2010,2011,2012],
      'team':['CSK','CSK','CSK','DDL','HSR','HSR','HSR','DDL'],
      'wins':[11,8,10,15,11,6,10,4],
      'losses':[5,8,6,1,5,10,6,12]
     
     }
print(data)
type(data)
```

```python
cricket=pd.DataFrame(data)
cricket
```

```python
# NOW Making another cttribute in DataFrame  called columns which let us arrange 
#our column according to we want

cricket=pd.DataFrame(data,columns=['year','team','wins','losses'])
print(cricket)
```

```python
cricket
```

### Reading : CSV

```python
from_csv =pd.read_csv('riverma02.csv',sep=',')
from_csv.head()
```

```python
 # we can be more explicit and define header =None to tell that there is no header is
    # in the dataset , if there will be a header then it will be treated as a row
    
from_csv =pd.read_csv('riverma02.csv',header=None ,sep=',')
from_csv.head()
```

### Addintionaly when we want to gice our own column name , then we can set  header=None and then pass array of column name to used as header by usinf 'names'

```python
cols = ['num', 'game', 'date', 'team', 'home_away', 'opponent',
        'result', 'quarter', 'distance', 'receiver', 'score_before',
        'score_after']

no_headers=pd.read_csv('peyton-passing-TDs-2012.csv',sep=',',header=None,names=cols)

no_headers.head()
```
