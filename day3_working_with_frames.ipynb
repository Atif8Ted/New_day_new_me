## Joining

````python
# like sql join clause , pandas.merge allows two dataframe to be joined in one or more keys
#The parameters are (on,left_on , right_on,left_index,right_index)
# By default pandas.merge provides inner join
# Although it can be changed to other joins usinf how paramete
# how:{ 'left','right','outer','inner'}, default 'inner'
````

### Good link for Joins in SQL:
http://www.sql-join.com/sql-join-types

## Joins: Concerpts
* left: use only keys from left frame (SQL : Left outer join) same as left join
* right: use only keys from right frame ( SQL: RIght outer join) same as right join
* Outer : use Union of keys from both frames ( SQL: Full outer join )  same as full join
* inner : use union of keys from both frames ( SQL: Inner join) 

## Basic examples of joins

````python
left_frame =pd.DataFrame({'key':range(5),'left_value':['a','b','c','d','e']})
right_frame=pd.DataFrame({'key':range(2,7),'right_value':['f','g','h','i','j']})
print(left_frame)
print("\n")
print(right_frame)
````

### Inner Join(Default)

````python
pd.merge(left_frame,right_frame,on='key',how='inner')
# Matching data i.e intersection got printed
````

### The sql equivalent to above will be
```sql
select left_frame.key,left_frame.left_value,right_frame.right_value from left_frame
INNER JOIN right_frame on left_frame.key=right_frame.key;

```

### If our key column were not same , we could have use left_on and right on parameter to specify which feilds to join from each frame

````python
# for example lets name keys column in both table differently 
left_frame =pd.DataFrame({'left_key':range(5),'left_value':['a','b','c','d','e']})
right_frame=pd.DataFrame({'right_key':range(2,7),'right_value':['f','g','h','i','j']})
print(left_frame)
print("\n")
print(right_frame)
````

````python
# Now lets call without using left_on parameter,it will give error
pd.merge(left_frame,right_frame,on='key',how='inner')
````

````python
# Now lets specify on which column to join from each frame usng left_on and right_on
pd.merge(left_frame,right_frame,left_on='left_key',right_on='right_key',how ='inner')
````
