## Joining

````python
# like sql join clause , pandas.merge allows two dataframe to be joined in one or more keys
#The parameters are (on,left_on , right_on,left_index,right_index)
# By default pandas.merge provides inner join
# Although it can be changed to other joins usinf how paramete
# how:{ 'left','right','outer','inner'}, default 'inner'
````

### Good link for Joins in SQL:
http://www.sql-join.com/sql-join-types

## Joins: Concerpts
* left: use only keys from left frame (SQL : Left outer join) same as left join
* right: use only keys from right frame ( SQL: RIght outer join) same as right join
* Outer : use Union of keys from both frames ( SQL: Full outer join )  same as full join
* inner : use union of keys from both frames ( SQL: Inner join) 

## Basic examples of joins

````python
left_frame =pd.DataFrame({'key':range(5),'left_value':['a','b','c','d','e']})
right_frame=pd.DataFrame({'key':range(2,7),'right_value':['f','g','h','i','j']})
print(left_frame)
print("\n")
print(right_frame)
````

### Inner Join(Default)

````python
pd.merge(left_frame,right_frame,on='key',how='inner')
# Matching data i.e intersection got printed
````

### The sql equivalent to above will be
```sql
select left_frame.key,left_frame.left_value,right_frame.right_value from left_frame
INNER JOIN right_frame on left_frame.key=right_frame.key;

```

### If our key column were not same , we could have use left_on and right on parameter to specify which feilds to join from each frame

````python
# for example lets name keys column in both table differently 
left_frame1 =pd.DataFrame({'left_key':range(5),'left_value':['a','b','c','d','e']})
right_frame1=pd.DataFrame({'right_key':range(2,7),'right_value':['f','g','h','i','j']})
print(left_frame)
print("\n")
print(right_frame)
````

````python
# Now lets call without using left_on parameter,it will give error
pd.merge(left_frame1,right_frame1,on='key',how='inner')
````

````python
# Now lets specify on which column to join from each frame usng left_on and right_on
pd.merge(left_frame1,right_frame1,left_on='left_key',right_on='right_key',how ='inner')
````

````python
# Alternatively , if our keys were indexex we could use the left_index or right_index ,
# which accepts a True/False value . You can mix and match columns and indices like so:
pd.merge(left_frame,right_frame,left_on='key',right_index=True)
````

### Left Outer Join or Left Join

````python
# it will include NULL or NaN , cause t is taking all the rows from left , that may be 
# or may not be present in the right table
pd.merge(left_frame,right_frame,on='key',how='left')
````

### SQL Equivalent for Left Join


```sql
select left_fram.key, left_frame.left_value ,right_frame.right_value from left_value left 
join right_frame on left_frame.key=right_frame.key;

```

### RIght outer Join or Right Join

````python
pd.merge(left_frame,right_frame,on='key',how='right')
````

### SQL EQUIVALENT for Right Join

```sql
select right_frame.key,left_frame.left_value,right_frame.right_value
from right_frame right join left_frame 
on right_frame.key=left_frame.key;
```

### Full Outer Join or Full Join

````python
pd.merge(left_frame,right_frame,on='key',how='outer')
````

### SQL Equivalent (though some databases don't allow FULL JOINs (e.g. MySQL)):

```sql
SELECT IFNULL(left_frame.key, right_frame.key) key
        , left_frame.left_value, right_frame.right_value
FROM left_frame
FULL OUTER JOIN right_frame
    ON left_frame.key = right_frame.key;
```



## Combining
* Pandas also provides a way to combine DataFraes alon an axis -"pandas.concat"
* While The function is equivalent to SQL, there is a lot more can be done with it .
* "pandas.concat" takes a 'list of series' or 'DataFrames'  and returns a Series or DataFrame of the concatnated objects .
* Because the funtion takes list , you can combine many objects at once.

````python
pd.concat([left_frame,right_frame])
````

By default, the function will vertically append the objects to one another, combining columns with the same name. We can see above that values not matching up will be NULL.

Additionally, objects can be concatentated side-by-side using the function's axis parameter.



````python
pd.concat([left_frame,right_frame],axis=1)
````

Since the data contains a dollar sign for each salary , python will treat the fields as series of string . We can use converters to change this when reading the file.
> Converters : dict.optional
* Dict of functions for converting values in certain columns . Keys can neither be integers or column




````python
headers=['name','title','department','salary']
chicago=pd.read_csv('city-of-chicago-salaries.csv',
                   header=0,
                   names=headers,
                   converters={'salary':lambda x: float(x.replace('$',''))})
chicago.head()
````

Pandas groupby returns a DataFrameGroupBy object which has a variety of methods,many of which are similar to standard SQL aggregate functions.


````python
by_dept=chicago.groupby('department')
by_dept
````

````python
#Calling "count" returns total no of NOT NULL values within each column
print(by_dept.count())# Not null records within each column 
print("\n")
print(by_dept.size()) # total records for each department
````

### Operations can also be done on an individual Series within a grouped object. Say we were curious about the five departments with the most distinct titles 

````python
by_dept.title.nunique().sort_values(ascending=False)[:5]
````

### Mysql Equivalent to :
```sql
select department,count (distinct title)
from chicago
group by department
order by desc
Limit 5;
```

````python
import numpy as np

def ranker(df):
    df['dept_rank']=np.arange(len(df))+1
    return df
````

````python
chicago.sort_values('salary',ascending=False ,inplace=True)
chicago=chicago.groupby('department').apply(ranker)
print(chicago[chicago.dept_rank==1].head(7))
````
